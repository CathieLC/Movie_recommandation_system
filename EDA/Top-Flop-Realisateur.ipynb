```python
# Charger la table "title_principals"

title_basics = pd.read_csv('https://datasets.imdbws.com/title.principals.tsv.gz', sep = '\t', index_col=0, low_memory=False)
```

```python
title_principals = title_basics 
```

```python
title_principals.info()
```

```python
# Merge avec inner pour garder seulement les données communes aux deux tables 

title_basics_plus_title_ratings_plus_title_principals = pd.merge(title_basics_plus_title_ratings, title_principals, how='inner', on= 'tconst')
```
